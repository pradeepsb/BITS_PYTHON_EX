```python
 # Ex1 - factorial of a number

def findFactorial(n):
  if(n<=0):
    return 0
  elif(n==1):
    return 1
  else:
    return n*findFactorial(n-1)

factorial = findFactorial(5)
print(factorial)
```

```python
#Ex2 - Given two lists, write a Python function to find the intersection (common elements) of the lists

def findIntersection(lst1, lst2):
  lst3 = []
  for value in lst1:
    if value in lst2:
      lst3.append(value)
  return lst3

lst1 = [1,2,4,6,-1,-3]
lst2 = [-1,3,9,10,6]
lst3 = findIntersection(lst1, lst2)
print(lst3)
```

```python
#Ex7 - Write a Python program to check if a given year is a leap year or not.
# Description: A year is a leap year if it is divisible by 4, except for years that are divisible by 100, unless they are also divisible by 400.

def findLeapYr(year1):
  ly = False

  if(year1 % 4 == 0 and year1 % 100 == 0 and year1 % 400 == 0):
    ly = True
  elif(year1 % 4 == 0 and year1 % 100 != 0):
    ly = True
  return ly

print(findLeapYr(1800))

```

```python
# Ex-10  Write a Python function that takes a list of integers and returns the sum of all the even numbers in the list.
#Example Input: [1, 2, 3, 4, 5]

def findSum(input):
  sum = 0
  for i in input:
    if(i % 2 == 0):
      sum = sum + i
  return sum

input = [1, 2, 3, 4, 5]
sum = findSum(input)

print(sum)
```

```python
#Ex-13: Write a Python function that takes a string and returns the length of the longest word in the string.

def findMaxWordLength(input):
  lst1 = input.split(" ")
  maxlength = 0

  for ls in lst1:
    if(len(ls) > maxlength):
      maxlength = len(ls)

  return maxlength

input = "The quick brown fox"

maxlength = findMaxWordLength(input)

print(maxlength)
```

```python
#Ex-15: Write a Python function that prints all prime numbers between start and end (inclusive).
# Example Input: start = 10, end = 20

def isPrimeNr(num):
  flag = True
  if(num > 1):
    for i in range(2, (num//2)+1):
      if (num % i) == 0:
        flag = False
  return flag

def PrintPrimeInrange(start, end):
  for i in range(start, end + 1):
    isPrime = isPrimeNr(i)
    if(isPrime):
      print(i)

start = 10
end = 20
PrintPrimeInrange(start, end)

```
